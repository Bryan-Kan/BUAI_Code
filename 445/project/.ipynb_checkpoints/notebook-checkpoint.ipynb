```python
# PerformanceMedia_Display
performanceMediaDisplayEffectiveness = datacsv['PerformanceMedia_Display_Leads'].sum()/datacsv['PerformanceMedia_Display_Visits'].sum()
print(performanceMediaDisplayEffectiveness)
```

```python
# PerformanceMedia_PaidSearch
performanceMediaPaidSearchEffectiveness = datacsv['PerformanceMedia_PaidSearch_Leads'].sum()/datacsv['PerformanceMedia_PaidSearch_Visits'].sum()
print(performanceMediaPaidSearchEffectiveness)
```

```python
# PerformanceMedia_Partner
performanceMediaPartnerEffectiveness = datacsv['PerformanceMedia_Partner_Leads'].sum()/datacsv['PerformanceMedia_Partner_Visits'].sum()
print(performanceMediaPartnerEffectiveness)
```

```python
df = pd.DataFrame({'Marketing Channel':['Display', 'PaidSearch', 'Partner'], 'value':[performanceMediaDisplayEffectiveness, performanceMediaPaidSearchEffectiveness, performanceMediaPartnerEffectiveness]})
ax = df.plot.bar(x='Marketing Channel', y='value', rot=0)
```
