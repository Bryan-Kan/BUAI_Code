## Q2  What is ROI from each channel?

```python
totalLeads = datacsv['PerformanceMedia_Display_Leads'].sum() + datacsv['PerformanceMedia_PaidSearch_Leads'].sum() + datacsv['PerformanceMedia_Partner_Leads'].sum()
print(totalLeads)
```

```python
# ROI = revenue / spend = revenue of each channel / spend of each channel
# Channel revenue = Total_DeseasonalizedMarketplace_Revenue * weight of each channel
weightMediaDisplay = datacsv['PerformanceMedia_Display_Leads'].sum() / totalLeads
MediaDisplayRevenue = datacsv['Total_DeseasonalizedMarketplace_Revenue'].sum() * weightMediaDisplay
performanceMediaDisplayROI = MediaDisplayRevenue / datacsv['PerformanceMedia_Display_Spends'].sum()
print(performanceMediaDisplayROI)
```

```python
weightMediaPaidSearch = datacsv['PerformanceMedia_PaidSearch_Leads'].sum() / totalLeads
MediaPaidSearchRevenue = datacsv['Total_DeseasonalizedMarketplace_Revenue'].sum() * weightMediaPaidSearch
performanceMediaPaidSearchROI = MediaPaidSearchRevenue / datacsv['PerformanceMedia_PaidSearch_Spends'].sum()
print(performanceMediaPaidSearchROI)
```

```python
weightMediaPartner = datacsv['PerformanceMedia_Partner_Leads'].sum() / totalLeads
MediaPartnerRevenue = datacsv['Total_DeseasonalizedMarketplace_Revenue'].sum() * weightMediaPartner
performanceMediaPartnerROI = MediaPartnerRevenue / datacsv['PerformanceMedia_Partners_Spends'].sum()
print(performanceMediaPartnerROI)
```

```python
df = pd.DataFrame({'Marketing Channel':['Display', 'PaidSearch', 'Partner'], 'value':[performanceMediaDisplayROI, performanceMediaPaidSearchROI, performanceMediaPartnerROI]})
ax = df.plot.bar(x='Marketing Channel', y='value', rot=0)
```
